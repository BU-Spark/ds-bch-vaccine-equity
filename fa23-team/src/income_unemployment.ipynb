```python
def convert_income_to_int(income_str):
    return int(income_str.replace(',', ''))
```

```python
wellesley_income_2020 = convert_income_to_int(dataframe_ws_2020["Wellesley CDP, Massachusetts"][13])
wellesley_income_2021 = convert_income_to_int(dataframe_ws_2021["Wellesley CDP, Massachusetts!!Median income (dollars)!!Estimate"][1])
wellesley_income_2022 = convert_income_to_int(dataframe_ws_2022["Wellesley CDP, Massachusetts!!Estimate"][0])

chelsea_income_2020 = convert_income_to_int(dataframe_ch_2020["Chelsea city, Massachusetts"][13])
chelsea_income_2021 = convert_income_to_int(dataframe_ch_2021["Unnamed: 5"][3])
chelsea_income_2022 = convert_income_to_int(dataframe_ch_2022["Chelsea city, Massachusetts"][1])

newbedford_income_2020 = convert_income_to_int(dataframe_nb_2020["New Bedford city, Massachusetts"][13])
newbedford_income_2021 = convert_income_to_int(dataframe_nb_2021["New Bedford city, Massachusetts"][13])
newbedford_income_2022 = convert_income_to_int(dataframe_nb_2022["New Bedford city, Massachusetts"][3])

revere_income_2020 = convert_income_to_int(dataframe_rv_2020["Revere city, Massachusetts"][13])
revere_income_2021 = convert_income_to_int(dataframe_rv_2021["Revere city, Massachusetts"][13])
revere_income_2022 = convert_income_to_int(dataframe_rv_2022["Revere city, Massachusetts"][1])

newton_income_2020 = convert_income_to_int(dataframe_nt_2020["Newton city, Massachusetts"][13])
newton_income_2021 = convert_income_to_int(dataframe_nt_2021["Newton city, Massachusetts"][13])
newton_income_2022 = convert_income_to_int(dataframe_nt_2022["Newton city, Massachusetts"][13])


```

```python
import requests
import json

# Your API key (replace with your actual API key)
api_key = 'YOUR_API_KEY'

# Base URL for the Census API
base_url = 'https://api.census.gov/data'

# Define the year and dataset
year = '2019'
dataset = 'acs/acs5'

# Define the parameters for the request
# Here, we are requesting the total population (B01001_001E) for all states
params = {
    'get': 'NAME,B01001_001E',
    'for': 'state:*',
    'key': api_key
}

# Construct the full URL
request_url = f'{base_url}/{year}/{dataset}'

# Make the request
response = requests.get(request_url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Load the response data
    data = json.loads(response.text)
    # Print the data
    for record in data:
        print(record)
else:
    print('Failed to retrieve data:', response.status_code)
```

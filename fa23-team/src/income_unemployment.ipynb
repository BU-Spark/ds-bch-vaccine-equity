```python
#Add Vaccination on The right y axis
vac_df = pd.read_csv('../data/final_data.csv')

vac_df['year_month'] = pd.to_datetime(vac_df['year_month'], format='%Y-%m')
vac_df['fully_vac_rate'].fillna(method='ffill', inplace=True)

vac_df.sort_values(by=['city', 'year_month'], inplace=True)

# Extract the 'year' from the 'year_month' column
vac_df['year'] = vac_df['year_month'].dt.year

# df_vaccination['fully_vac_rate'].fillna(method='ffill', inplace=True)

# Group the data by 'city' and 'year', and calculate the mean vaccination rate for each group
average_yearly_vaccination_rate = vac_df.groupby(['city', 'year'])['fully_vac_rate'].mean().reset_index()
```

## Unemployment Rate Monthly FROM 2020-2023

```python
import pandas as pd

df = pd.read_csv('../data/unemployment_2020-2023.csv')

print(df.head())
```

```python
# Convert 'Month' to a numerical format
months = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6, 'July':7, 'August': 8, 'September':9,
         'October':10, 'November':11, 'December': 12}
df['Month'] = df['Month'].map(months)
df = df.dropna(subset=['Month'])
df['Month'] = df['Month'].astype(int)

# Sort by 'Year' and then by 'Month'
sorted_df = df.sort_values(by=['Year', 'Month'])
```

```python
sorted_df['Year'] = sorted_df['Year'].astype(str)
sorted_df['Month'] = sorted_df['Month'].apply(lambda x: f'{x:02d}')  # Formats the month as a two-digit number


sorted_df['year-month'] = sorted_df['Year'] + '-' + sorted_df['Month']
```

```python
sorted_df
```

```python
mask = (sorted_df['year-month'] >= '2021-03') & (sorted_df['year-month'] <= '2023-04')
filtered_df = sorted_df[mask]
```

```python
filtered_df = filtered_df.rename(columns={'Area': 'city', 'Area Rate': 'city_rate'})
```

```python
import pandas as pd

# Save DataFrame to CSV file
filtered_df.to_csv('../data/unemployement_filtered.csv', index=False)  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the list of cities and exotic colors
cities_to_plot = ["New Bedford", "Chelsea", "Revere", "Newton", "Wellesley"]
exotic_colors = ['#d01d20', '#ff6f18', '#FFD700', '#2fad98', '#1e52b8']

# Filter the DataFrame to include only the desired cities
colored_df = filtered_df[filtered_df['city'].isin(cities_to_plot)]

# Create a custom color palette mapping each city to its corresponding color
custom_palette = {city: color for city, color in zip(cities_to_plot, exotic_colors)}

plt.figure(figsize=(20, 10))
filtered_df['year-month'] = filtered_df['year-month'].astype(str)

# Set the custom color palette
sns.set_palette([custom_palette[city] for city in filtered_df['city']])

# Plot the lineplot
sns.lineplot(data=filtered_df, x='year-month', y='city_rate', hue='city', marker='o', ci=None)

# Adjusting x-axis for better readability
plt.xticks(rotation=30)

# Set title and labels
plt.title('Unemployment Rate over Time for Selected Cities', fontsize=25)
plt.ylabel('Unemployment Rate (%)', fontsize=20)
plt.xlabel('Year-Month', fontsize=20)

# Adjust the tick labels and legend
plt.xticks(rotation=30, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(title='Cities', title_fontsize=17, fontsize=20, loc='upper right')

plt.tight_layout()
plt.show()
```
